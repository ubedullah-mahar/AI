```python
import pandas as pd

student_data = {
    'Student_ID': [201, 202, 203, 204, 205],
    'Name': ['Ayesha', 'Bilal', 'Danish', 'Hina', 'Umar'],
    'Department': ['CS', 'IT', 'CS', 'SE', 'IT'],
    'Marks': [86, 74, 91, 68, 88],
    'Attendance': [90, 82, 96, 78, 92]
}

df = pd.DataFrame(student_data)

# 1. Displaying only students from  IT department
it_students = df[df['Department'] == 'IT']
print("\n1. Students from IT Department:")
print(it_students)
```

```python
# 2. Calculating the average marks in each department
avg_marks_by_dept = df.groupby('Department')['Marks'].mean().reset_index()
avg_marks_by_dept.columns = ['Department', 'Average_Marks']
print("\n Average Marks by Department")
print(avg_marks_by_dept)
```

```python
# 3. Addinc new column Status
df['Status'] = df['Marks'].apply(lambda x: 'Pass' if x >= 70 else 'Fail')
print(df)
```

```python
# 4. Students with attendance less than 85%
low_attendance = df[df['Attendance'] < 85]
print(low_attendance)
```

```python
# 5. Sorting the DataFrame by Marks (highest to lowest)
df_sorted = df.sort_values(by='Marks', ascending=False)
print(df_sorted)
```
